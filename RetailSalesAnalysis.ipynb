```python
# Create a total sales column, quantity X Price = Revenue per row
df["TotalSales"]=df["QUANTITYORDERED"] * df["PRICEEACH"]
```

```python
# I would like to look at a few things such as the order dates, quantity, prices, total sales.

df[["QUANTITYORDERED", "PRICEEACH", "TotalSales"]].head()
```

```python
# Perform an aggregation to sum all the sales for each category
sales_by_category = (
    df.groupby("PRODUCTLINE")["TotalSales"]
    .sum()
    .sort_values(ascending=False)
)

sales_by_category
```

```python
# Another aggregation for the monthly sales
# Extract the month
df["YearMonth"]=df["ORDERDATE"].dt.to_period("M")

monthly_sales = df.groupby("YearMonth")["TotalSales"].sum()
monthly_sales
```

```python
# Finding top products
top_products = (
    df.groupby("PRODUCTLINE")["TotalSales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_products
```

```python
# From this analysis I found that the largest share of revenue comes from the Classic cars category followed by Vintage cars.
# Found a pattern of sales steadily rising in October and hitting the peak revenue in November.
# Top revenue is concentrated in the cars category.
```

```python
sales_by_category.to_csv("../outputs/sales_by_category.csv")
monthly_sales.to_csv("../outputs/monthly_sales.csv")
top_products.to_csv("../outputs/top_products.csv")
```
